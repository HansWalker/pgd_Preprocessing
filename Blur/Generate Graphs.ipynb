```python
import numpy as np
import matplotlib.pyplot as plt
image_dir='mnist/loop_data/'
carray=np.load(image_dir+'mnist_loop_list_corr0.npy')
barray=np.load(image_dir+'mnist_loop_list0.npy')
carray=np.zeros([len(carray),len(carray[0])])
barray=carray.copy()
listnum=75
for i in range(listnum):
    carray+=np.load(image_dir+'mnist_loop_list_corr'+str(i)+'.npy')
    barray+=np.load(image_dir+'mnist_loop_list'+str(i)+'.npy')
carray=carray/listnum
barray=barray/listnum
fig, ax = plt.subplots()
ax.plot(carray[0],carray[1])
ax.plot(carray[0],barray[1])
```
